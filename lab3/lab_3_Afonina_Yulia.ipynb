```python
#  Активируем библиотеки
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn import preprocessing
matplotlib.style.use('ggplot')
%matplotlib inline
import os
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
#  Активируем процедуру DBSCAN
from sklearn.cluster import DBSCAN
```

```python
#  Задаем рабочую папку
os.chdir("/content/drive/MyDrive/ITMO/data analysis/lab_3")
```

```python
#  Указываем колонку City как index_col
df = pd.read_csv('/content/drive/MyDrive/ITMO/data analysis/lab_1/Econom_Cities_data.csv', sep = ';', decimal=',',  index_col='City')
```

```python
# Удалим выбросы
```

```python
print(df[df.index=='Cairo'])
print(df[df.index=='Jakarta'])

df = df.drop('Jakarta');
df = df.drop('Cairo');
```

```python
df
```

```python
# Данные находятся в разных диапазонах. Нужна стандартизация
# z = (x - mean) / std - формула стандартизации
```

```python
stand_1 = preprocessing.StandardScaler() # Создаем объект scaler
stand_1.fit(df) # Изучаем данные (вычисляем среднее и std)
X = stand_1.transform(df)  # Применяем преобразование
X = pd.DataFrame(X, index=df.index, columns=df.columns)
```

```python
# Кластеризуем методом DBSCAN
# инициализируем модель
dbscan = DBSCAN(eps=0.5, metric='euclidean', min_samples=5)
# обучаем модель
dbscan.fit(X)
# смотрим на результаты кластеризации
dbscan.labels_
```

```python
# Представляем результаты кластеризации
# Создаем таблицу частот в numpy
unique, counts = np.unique(dbscan.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
```

```python
# Видим, что очень много образовалось значений -1 (36 штук)
# Это говорит о потенциальных выбросах, но ввиду того,
# что выборка небольшая, их чересчур много
# Попробуем увеличивать eps
```

```python
# Снова инициализируем модель, но уже с параметром eps=5
dbscan = DBSCAN(eps=5, metric='euclidean', min_samples=5)
# обучаем модель
dbscan.fit(X)
# смотрим на результаты кластеризации
dbscan.labels_
```

```python
# Получили один кластер
# Попробуем уменьшить eps в 2 раза
# Также попробуем уменьшить min_samples до 3
```

```python
# инициализируем модель
dbscan = DBSCAN(eps=2.5, metric='euclidean', min_samples=3)
# обучим модель
dbscan.fit(X)
# смотрим на результаты кластеризации
dbscan.labels_
```

```python
# Аналогично. Попробуем уменьшить min_samples до 2, а eps до 1.2.
```

```python
# инициализируем модель
dbscan = DBSCAN(eps=1.2, metric='euclidean', min_samples=2)
# обучим модель
dbscan.fit(X)

unique, counts = np.unique(dbscan.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
```

```python
# Видим, что выделилось 2 кластера, содержащие 42 и 2 объекта,
# а так же 2 значения выброса
# Количество значений, входящих в кластеры, неравномерно, попробуем ещё
```

```python
# инициализируем модель
dbscan = DBSCAN(eps=0.8, metric='euclidean', min_samples=2)
# обучим модель
dbscan.fit(X)

unique, counts = np.unique(dbscan.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
```

```python
# Остановимся на этом решении
# Дальнейшее уменьшение min_samples приводит к излишней раздробленности кластеров
# А дальнейшее уменьшение eps - к увеличению выбросов
# Увеличение eps приводит к соединению кластеров 0 и 1 в один кластер, содержащий 38 объектов.
# Количество объектов в кластерах в текущем решении неравномерно
# Думаю, можно это объяснить несбалансированностью обучающей выборки
```

```python
df['dbscan'] = dbscan.labels_
df['dbscan'].sort_values()
```

```python
df.groupby('dbscan').mean()
```

```python
# В кластере 0 - в меру работают и в меру получают.

# В кластере 1 - самые плохо оплачиваемые города - часов много при низких зарплатах.

# В кластере 2 - самое лучшее соотношение между часами и полученной зарплатой
# (зарплаты тут выше всего).

# В кластере 3 при меньшем количестве часов наблюдаются хорошие зарплаты,
# но при этом высокие цены.

# Кластер -1 можно посчитать выбросами. Либо же охарактеризовать так - люди много работают,
# получают средне, но при этом цены высокие.

# Вывод: конкретно в данной задаче гораздо удобнее было применить
# иерархический кластерный анализ или же K-means.
# Подбирать eps и num_samples не всегда очевидная и тривиальная задача.
```
